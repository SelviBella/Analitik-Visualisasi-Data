# Business Understanding

## Business Objective
Dataset ini berisi informasi produk parfum yang dijual pada platform e-commerce Noon. Data mencakup berbagai atribut seperti nama produk, brand, kategori, target pengguna, harga, dan rating. Dalam industri e-commerce, harga dan kualitas produk (yang direpresentasikan melalui rating) merupakan faktor penting yang memengaruhi keputusan pembelian konsumen. Oleh karena itu, analisis terhadap distribusi harga, perbedaan antar brand, serta hubungan antara harga dan rating menjadi penting untuk memahami pola pasar parfum di platform tersebut.

Adapun tujuan dari analisis bisnis ini adalah:

1. Mengidentifikasi distribusi harga produk parfum.
2. Menganalisis perbedaan harga berdasarkan brand dan kategori.
3. Mengetahui hubungan antara harga dan rating produk.
4. Mengidentifikasi segmen parfum dengan rating tertinggi.
5. Memberikan insight mengenai posisi pasar brand parfum di platform Noon.

## Assess Situation
Sebelum melakukan analisis, kita perlu memahami kondisi dataset dan juga tantangannya. Dikarenakan data diambil melalui web scraping, kemungkinan ada beberapa masalah seperti data kosong (missing values), penulisan brand yang tidak konsisten, atau data yang duplikat. Selain itu, bisa saja terdapat harga yang terlalu tinggi atau terlalu rendah (outlier) yang dapat memengaruhi hasil analisis. Dataset ini juga hanya berasal dari satu platform yaitu Noon, sehingga hasil analisis hanya menggambarkan kondisi di platform tersebut.

## Analytic Goals
Berdasarkan tujuan bisnis yang telah ditentukan, maka tujuan analitik yang akan dilakukan adalah:

• Menganalisis distribusi harga produk parfum.

• Membandingkan harga berdasarkan brand dan kategori produk.

• Mengidentifikasi hubungan antara harga dan rating menggunakan analisis korelasi atau visualisasi scatter plot.

• Mengidentifikasi brand atau segmen produk dengan performa terbaik berdasarkan rating dan rentang harga.

Tujuan analitik ini dirancang untuk menjawab permasalahan bisnis secara lebih teknis dan terukur melalui pendekatan Exploratory Data Analysis (EDA).
